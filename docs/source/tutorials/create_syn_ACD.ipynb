```python
from pgm.synthetic.syn_acd import SyntNetAnomaly

syn_acd_net = SyntNetAnomaly(
    N=300,
    K=3,
    avg_degree=15,
    verbose=2,
    structure="assortative",
    ag=1.5,
    bg=10,
    rho_anomaly=0.1,
    output_adj=True,
    outfile_adj="network_with_anomalies.csv",
)
```

```python
networks = syn_acd_net.anomaly_network_PB()
```

```python
networks
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph from the first adjacency matrix in the list A
graph = networks[0]

# Get the adjacency matrix
adjacency_matrix = nx.adjacency_matrix(graph)

# Plot the adjacency matrix
plt.imshow(adjacency_matrix.toarray(), cmap="Greys", interpolation="none")

# Add a color bar to the plot for reference
plt.colorbar()
plt.show()
```
